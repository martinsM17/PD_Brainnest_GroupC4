4. A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 Ã— 99.

Find the largest palindrome made from the product of two 3-digit numbers.


```python
largest_palindrome = 0 # initialize largest_palindrome to 0

# Loop through all pairs of three-digit numbers (i and j)
for i in range(100, 1000):
    for j in range(100, 1000):
        product = i * j # calculate the product of i and j
        if str(product) == str(product)[::-1]: # check if the product is a palindrome
            largest_palindrome = max(largest_palindrome, product) # update largest_palindrome if the product is larger

# Print the largest palindrome
print(largest_palindrome)
```

5. 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?


```python
primes = [2,3,5,7,11,13,17,19] # prime factors of each number from 1 to 20

prod = 1 # initialize product to 1

# Loop through each prime factor
for p in primes:
    n = 2 # start with the second power of the prime factor
    prod *= p # multiply the product by the prime factor
    while (p**n < 21): # loop through powers of the prime factor up to 20
        prod *= p # multiply the product by the prime factor raised to the current power
        n += 1 # increment the power

# Print the final product
print(prod)
```

8. A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
a^2 + b^2 = c^2

For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.


```python
def pit_trip():
    for i in range(1, 1000):
        for j in range(i, 1000):
            h = 1000 - i - j
            if h > 0:
                if i**2+j**2 == h**2:
                    return int(i * j * h)
    return None
result = pit_trip()
print(result)
```

9. The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.


```python
def nth_prime(n):
    primes = []
    count = 0
    i = 2
    while True:
        is_prime = True
        for p in primes:
            if i % p == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(i)
            count += 1
            if count == n:
                return p
        i += 1
result = nth_prime(10001)
print(result) 
```
